## Adding Family Size Column

```python
df['Family_Size_Number'] = df['SibSp'] + df['Parch'] + 1
df.sample(5)
```

```python
def family_size(size):
    if size == 1:
        return 'Alone'
    elif size > 1 and size < 5:
        return 'Small'
    else:
        return 'Large'
```

```python
df['Family_Size'] = df['Family_Size_Number'].apply(family_size)
df
```

```python
df['Family_Size'].value_counts()
```

```python
df['Family_Size'].value_counts().plot(kind='bar')
```

```python
df['Family_Size'].value_counts().plot(kind='pie', autopct='%0.1f%%')
```

## Bivariate Analysis

### Survived and Pclass

#### Conclusions
- Highest probability of death -> Pclass 3
- Lowest probability of death -> Pclass 1

```python
pd.crosstab(df['Survived'], df['Pclass'], normalize='columns')*100
```

### Survived and Sex
- Females had a much higher chance of survival than males

```python
pd.crosstab(df['Survived'], df['Sex'], normalize='columns')*100
```

### Embarked

### Survived and Age
- Ages 0-10 had the higher chances of survival
- Ages 20-40 had lower chances of survival

```python
df[df['Survived'] == 1]['Age'].plot(kind='kde', label='Survived')
df[df['Survived'] == 0]['Age'].plot(kind='kde', label='Not Survived')

plt.legend()
```

### Survived and Family Size
- Travelling in a small family had the highest chances of survival
- Travelling in a large family had the lowest chances of survival

```python
pd.crosstab(df['Survived'], df['Family_Size'], normalize='columns')*100
```

```python
df[df['Survived'] == 1]['Family_Size_Number'].plot(kind='kde', label='Survived')
df[df['Survived'] == 0]['Family_Size_Number'].plot(kind='kde', label='Not Survived')

plt.legend()
```
